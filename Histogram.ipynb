```python
def Outlier(a):
    Q1=a.quantile(0.25)
    Q3=a.quantile(0.75)
    IQR=Q3-Q1
    In=(Q1-(1.5*IQR))
    Out=(Q3+(1.5*IQR))
    print('1st Quantile =',Q1,'\n3rd Quantile =',Q3,'\nInner_Fence =',In,'\nOuter_Fence =',Out)
    for i in a:
        if (i<In or i>Out):
            print (i)
```

```python
Outlier(df.mpg)
```
