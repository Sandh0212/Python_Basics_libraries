```python
def impute_GRE(cols):
    GRE = cols[0]
    Admit=cols[1]
    if pd.isnull(GRE):
        if Admit==0:
            return 308
    elif Admit==1:
        return 323
    else:
        return GRE
```

```python
df['GRE']=df[['GRE','Admit']].apply(impute_GRE,axis=1)
```
