```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
plt.plot(tips['total_bill'],tips['tip'],'o')
```

```python
fig =plt.figure()
axes=fig.add_subplot(1,1,1)
axes.hist(tips['total_bill'],bins=8)
axes.set_title('Distribution of total bill')
```

```python
boxplot=plt.figure()
axes=boxplot.add_subplot(1,1,1)
axes.boxplot([tips[tips['sex']=='Female']['tip'],tips[tips['sex']=='Male']['tip']],labels=['Female','Male'])
axes.set_xlabel('Gender')
axes.set_ylabel('Tip')
axes.set_title('BoxPlot')
boxplot.show
```

```python
fig =plt.figure()
axes=fig.add_subplot(1,1,1)
axes.hist(tips[tips['sex']=='Female']['tip'],bins=8)
axes.set_title('Distribution of Female tip')
```

```python
fig =plt.figure()
axes=fig.add_subplot(1,1,1)
axes.hist(tips[tips['sex']=='Male']['tip'],bins=8)
axes.set_title('Distribution of Male tip')
```

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1 
```

```python
tips['gender_colour']=tips['sex'].apply(recode_sex)
```

```python
tips
```

```python
scatter_plot=plt.figure()
axes=scatter_plot.add_subplot(1,1,1)
axes.scatter(x=tips['total_bill'],y=tips['tip'],c=tips['gender_colour'],alpha=0.8)
axes.set_title('Scatter Plot')
axes.set_xlabel('Total Bill')
axes.set_ylabel('Tip Given')
```

```python
ax=sns.distplot(tips['total_bill'])
ax.set_title('Histogram with distribution')
```

```python
ax=sns.distplot(tips[tips['sex']=='Female']['total_bill'])
ax.set_title('Histogram with distribution')
```

```python
ax=sns.distplot(tips[tips['sex']=='Male']['total_bill'])
ax.set_title('Histogram with distribution')
```

```python
den,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],hist=False)
ax.set_title('Total bill density')
ax.set_xlabel('Total bill')
ax.set_ylabel('Unit Probability')
```

```python
den,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total bill density')
ax.set_xlabel('Total bill')
ax.set_ylabel('Unit Probability')
```

Rug plots are one dimensional representation of a variable's distribution

```python
hist_den_rug,ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title('Total bill histogram with density and rug')
ax.set_xlabel('Total Bill')
```

```python
count,ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title('Bar Plot')
ax.set_xlabel('DAY of the week')
ax.set_ylabel('Count or frequency')
```

```python
scatter,ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Scatter plot with regression line')
ax.set_xlabel('Total bill')
ax.set_ylabel('Tip given')
```

```python
fig=sns.lmplot(x='total_bill',y='tip',data=tips)
```
